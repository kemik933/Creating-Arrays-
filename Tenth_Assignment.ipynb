```python
# EXERCISE 3.1: Given an array, extract:
# 1. All elements greater than the mean
# 2. All odd numbers
# 3. Elements at specific positions

# Write your code here:

Array = [15, 20, 22, 23, 33, 45, 60, 75, 82, 85]

# 1. Greater than mean
Mean_Value = sum(Array) / len(Array)
greater_than_mean = [x for x in Array if x > Mean_Value]
print("Mean:", Mean_Value)
print("Greater than mean:", greater_than_mean)

# 2. Odd numbers
odd_numbers = [x for x in Array if x % 2 != 0]
print("Odd numbers:", odd_numbers)


# 3. Elements at specific positions
positions = [0, 2, 4]
elements_at_positions = [Array[i] for i in positions]
print("Elements at positions:", elements_at_positions)
```

```python
# EXERCISE 4.1: Create two arrays and perform:
# 1. Element-wise multiplication
# 2. Add scalar value to all elements
# 3. Calculate square roots

# Write your code here:

import numpy as np

# Create two arrays
A = np.array([25, 81, 49, 144])
B = np.array([1, 3, 5, 7])

# 1. Element-wise multiplication
elementwise_product = A * B
print("Element-wise multiplication:", elementwise_product)

# 2. Add scalar
Scalar_Added = A + 10
print("Add scalar:", Scalar_Added)

# 3. Square roots
Sqrt_Values = np.sqrt(A)
print("Square roots:", Sqrt_Values)
```
