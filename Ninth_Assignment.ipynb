# EXERCISE 2.1: Create the following arrays:
# 1. Array of 10 zeros
# 2. Array from 0 to 100 with step 10
# 3. 5 evenly spaced numbers between 0 and 1

# Write your code here:

```python
import numpy as np
```

```python
# 1. Array of 10 zeros

Array_1 = np.zeros(10)
print("Array of 10 zeros:\n", Array_1)
```

```python
# 2. Array from 0 to 100 with step 10

Array_2 = np.arange(0, 101, 10)
print("\nArray from 0 to 100 with step 10:\n", Array_2)
```

```python
# 3. 5 evenly spaced numbers between 0 and 1
Array_3 = np.linspace(0, 1, 5)
print("\n5 evenly spaced numbers between 0 and 1:\n", Array_3)
```
